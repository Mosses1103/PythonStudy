## Numpy란?

Numpy는 C언어로 구현된 파이썬 라이브러리로써, 고성능의 수치계산을 위해 제작되었습니다. Numerical Python의 줄임말이기도 한 Numpy는 벡터 및 행렬 연산에 있어서 매우 편리한 기능을 제공합니다.

## Pandas vs Numpy

주로 Pandas로 DataFrame을 만들고, 안에 데이터를 담을 때 Numpy를 사용

## Numpy_Array_Math

```python
data1 = [[1,2,3]
         ,[4,5,6]
         ,[97,98,99]]
x = np.array(data1)
print(x)

data2 = [[10,11,12]
         ,[14,15,16]
         ,[17,18,19]]
y = np.array(data2)
print(y)
```

```python
z1 = x + y
z2 = np.add(x,y)
print(z1)
print(z2)
```

```python
z1 = x - y
z2 = np.subtract(x,y)
print(z1)
print(z2)
```

```python
z1 = x * y
z2 = np.multiply(x,y)
print(z1)
print(z2)
```

```python
z1 = x / y
z2 = np.divide(x,y)
print(z1)
print(z2)
```

```python
z1 = x // y # 소수점 버림
z2 = np.floor(np.divide(x,y))
z3 = np.floor_divide(x,y)
print(z1)
print(z2)
print(z3)
```

```python
data1 = [[1,2]
         ,[4,5]]
x = np.array(data1)
print(x)

data2 = [[10,11]
         ,[14,15]]
y = np.array(data2)
print(y)
```

## Numpy_Array_Broadcasting (배열확산변환)

```python
data1 = [[1,2,3,4] # 0 행, 4열
         ,[5,6,7,8] # 1 행 , 4열
         ,[9,10,11,12]] # 2 행, 4열
data2 = [1,0,1,0]
x = np.array(data1)
y = np.array(data2)
z = np.zeros_like(x)
print("x :", x)
print("y :", y)
print("z :", z)
```

```python
for i in range(3):
    z[i, :] = x[i, :] + y # 행 더하기
    print("x", i , "행 + y", i, "행")
print(z)
```

```python
y1 = [1,0,1]
for i in range(4):
    z[:, i] = x[:, i] + y1 # 열 더하기
    print("x", i , "열 + y", i, "")
print(z)
```

```python
y2 = np.tile(y, (3, 1))
print("y :", y)
print("y2 :", y2)
```

```python
z = x + y2
print(z)
```

```python
z = x + y # 자동 확산
print(z)
```

```python
print(x[0,:])
print(x[0,:].shape)
print(y)
print(y.shape)
```

```python
z = np.reshape(x[0,:],(4,1))*y
print(z)
```

# 3주차 Matplotlib

## Matplotlib란?

Matplotlib는 파이썬에서 자료를 차트(chart)나 플롯(plot)으로 시각화(visulaization)하는 패키지이다.

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline # 브라우저에서 바로 그림을 볼 수 있게 해주는 것
```

```python
x = np.arange(0, 3 * np.pi, 1) # 0 ~ 3*pi(9.xx)
y = np.sin(x)

plt.plot(x, y)
```

```python
x = np.arange(0, 3 * np.pi, 0.1) # 0 ~ 3*pi(9.xx)
y = np.sin(x)

plt.plot(x, y)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.rcParams["figure.figsize"] = (10,5) # 그림(figure)의 크기. (가로,세로) 인치 단위
plt.rcParams['lines.linewidth'] = 5 # 선의 두께
plt.rcParams['axes.grid'] = True  # 차트내 격자선(grid) 표시 여부, 적용이 한템포 느리다
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Title : Sine and Cosine')
plt.legend(['1_sin', '2_cos'])
```

```python
plt.rcParams['lines.linewidth'] = 1 # 선의 두께

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)

plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

plt.show()
```

```python
plt.rc("lines", linewidth=2, color='g')

fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(x, y_sin)
axes[1].plot(x, y_cos)
axes[0].set_title('Sine')
axes[1].set_title('Cosine')

plt.show(fig)
```
